1. Completa el código para calcular el histograma de 1D, con solo 2 colores

```python
u= np.random.randint(2, size=12)
print(u)
```

```python
histograma = np.zeros(2, dtype= np.int32)

for i in range( len(u) ):
    if u[  i] == 0  :
      histograma[ 0 ] +=1
    else:
      histograma[ 1 ] +=1

print(histograma)
```

2. Modifica el codigo para que acepte 10 colores

```python
v= np.random.randint(10, size=12)
print(v)
```

```python
histograma10 = np.zeros(10, dtype= np.int32)

for i in range( len(v) ):
    ind = v[ i ]
    histograma10[ ind  ] += 1

print(histograma10)
```

3. Aplica el código anterior a una imagen 2D, con 10 Colores

```python
Im= np.random.randint(5, size=(3,6))
A=Im.shape
print(A)
print(Im)
```

```python
hist= np.zeros(10, dtype= np.int32)
A=Im.shape
M,N = A

for i in range( M  ):
  for j in range( N  ):
    ind = Im[ i   , j   ]
    hist[ ind ] += 1

print(hist)
```

```python
Im1 = io.imread('pexels-pixabay-220938.jpg')[:,:,0]
Im2 = io.imread('OIP.webp')[:,:,0]
Im3 = io.imread('OIP (1).webp')[:,:,0]
Im4 = io.imread('poster.jpg',)[:,:,0]
```

```python
imshow(Im1)
imshow(Im2)
imshow(Im3)
imshow(Im4)
```

Elige una de las imagenes anteriores y asignla a la imagen Im

```python
Im= Im4
Im.shape
```

Calcula el histograma de la imagen Im

```python
histograma = np.zeros(256, dtype= np.int32)

M,N= Im.shape

for i in range(M   ):
  for j in range( N  ):
    ind= Im[ i , j ]
    histograma[ ind  ] +=1

print(histograma)
```
